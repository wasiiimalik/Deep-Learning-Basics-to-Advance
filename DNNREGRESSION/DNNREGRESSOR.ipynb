```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("housing.csv")
df_data = data.select_dtypes(include=[np.float,np.int])
data = df_data.drop(['LotFrontage','MasVnrArea','GarageYrBlt','Id'],axis = 1)
```

```python
data.info()
```

```python
X = data.drop('SalePrice',axis = 1)
```

```python
Y = data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.20,random_state = 40)
```

```python
X_train = X_train.values
```

```python
X_test = X_test.values
```

```python
y_train = y_train.values
```

```python
y_train = y_train.values
```

```python
y_test = y_test.values
```

```python
X_train
```

```python
X_test
```

```python
model = Sequential()
model.add(Dense(50, input_shape=(33,), activation='relu'))
model.add(Dense(256,  activation='relu'))
model.add(Dense(562,  activation='relu'))
model.add(Dense(256,  activation='relu'))
model.add(Dense(256,  activation='relu'))
model.add(Dense(50,  activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
model.compile(loss='mean_absolute_error', optimizer='adam')
model.summary()
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=32, validation_split = 0.2)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(f"accuracy is {r2_score(y_test,predictions) * 100} ")
```
