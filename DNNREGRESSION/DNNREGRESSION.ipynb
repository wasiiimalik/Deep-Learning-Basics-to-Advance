```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
df_data = data.select_dtypes(include=[np.float,np.int])
```

```python
df_data.info()  
```

```python
data.isnull().sum()
```

```python
data = df_data.drop(['LotFrontage','MasVnrArea','GarageYrBlt','Id'],axis = 1)
```

```python
data.isnull().sum()
```

```python
data
```

```python
dataarray = data.values
```

```python
dataarray
```

```python
X = dataarray[:,0:33] #indepen var
Y = dataarray[:,-1] #dependent
```

```python
Y
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(33,), activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
model.compile(loss='mean_absolute_error', optimizer='adam')
model.summary()
```

```python
model.fit(X, Y, epochs=20, batch_size=32, validation_split = 0.2)
```

```python
predictions = model.predict(X)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
print(f"accuracy is {r2_score()}")
```
